```python
# import pandas
import pandas as pd

# Read the csv file into a variable insurance_data
insurance_data = pd.read_csv("/kaggle/input/insurance/insurance.csv")
insurance_data.head()
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Plot a scatter plot for bmi against the insurance charges
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

It is positively correlated

```python
# Add a regression line to show what best fits the data
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#Colour code the scatter plot with points by 'smoker'
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
#Add 2 regression lines corresponding to smokers and non-smokers
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
# For categorical scatter plot
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

non-smokers are less charged than smokers on insurance
